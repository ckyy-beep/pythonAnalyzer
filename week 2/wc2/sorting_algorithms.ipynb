```python
import random

# Maak drie arrays met 20 willekeurige integers
array1 = [random.randint(0, 50) for i in range(20)]
array2 = [random.randint(0, 50) for i in range(20)]
array3 = [random.randint(0, 200) for i in range(20)]

# Print de arrays
print("Array 1:", array1)
print("Array 2:", array2)
print("Array 3:", array3)

print()

def bubble_sort(array):
  for i in range(len(array) - 1):
    for j in range(len(array) - i - 1):
      if array[j] > array[j + 1]:
        array[j], array[j + 1] = array[j + 1], array[j]

  return array


def selection_sort(array):
  for i in range(len(array)):
    min_index = i
    for j in range(i + 1, len(array)):
      if array[j] < array[min_index]:
        min_index = j

    array[i], array[min_index] = array[min_index], array[i]

  return array


def insertion_sort(array):
  for i in range(1, len(array)):
    current_element = array[i]
    j = i - 1

    while j >= 0 and array[j] > current_element:
      array[j + 1] = array[j]
      j -= 1

    array[j + 1] = current_element

  return array

def merge_sort(array):
    if len(array) <= 1:
        return array

    mid = len(array) // 2
    left_half = merge_sort(array[:mid])
    right_half = merge_sort(array[mid:])

    return merge(left_half, right_half)


def merge(left_array, right_array):
    merged_array = []
    i = 0
    j = 0

    while i < len(left_array) and j < len(right_array):
        if left_array[i] <= right_array[j]:
            merged_array.append(left_array[i])
            i += 1
        else:
            merged_array.append(right_array[j])
            j += 1

    merged_array.extend(left_array[i:])
    merged_array.extend(right_array[j:])

    return merged_array


def quick_sort(array):
    if len(array) <= 1:
        return array

    pivot = array[0]
    left_array = [element for element in array[1:] if element <= pivot]
    right_array = [element for element in array[1:] if element > pivot]

    return quick_sort(left_array) + [pivot] + quick_sort(right_array)


# Sorteer de arrays met de verschillende algoritmes
sorted_array1_bubble = bubble_sort(array1.copy())
sorted_array1_selection = selection_sort(array1.copy())
sorted_array1_insertion = insertion_sort(array1.copy())
sorted_array1_merge = merge_sort(array1.copy())
sorted_array1_quick = quick_sort(array1.copy())

sorted_array2_bubble = bubble_sort(array2.copy())
sorted_array2_selection = selection_sort(array2.copy())
sorted_array2_insertion = insertion_sort(array2.copy())
sorted_array2_merge = merge_sort(array2.copy())
sorted_array2_quick = quick_sort(array2.copy())

sorted_array3_bubble = bubble_sort(array3.copy())
sorted_array3_selection = selection_sort(array3.copy())
sorted_array3_insertion = insertion_sort(array3.copy())
sorted_array3_merge = merge_sort(array3.copy())
sorted_array3_quick = quick_sort(array3.copy())

# Print de gesorteerde arrays
# print("Array 1 na bubble sort:", sorted_array1_bubble)
# print("Array 1 na selection sort:", sorted_array1_selection)
# print("Array 1 na insertion sort:", sorted_array1_insertion)
# print("Array 1 na merge sort:", sorted_array1_merge)
# print("Array 1 na quick sort:", sorted_array1_quick)

print()

# print("Array 2 na bubble sort:", sorted_array2_bubble)
# print("Array 2 na selection sort:", sorted_array2_selection)
# print("Array 2 na insertion sort:", sorted_array2_insertion)
# print("Array 2 na merge sort:", sorted_array2_merge)
# print("Array 2 na quick sort:", sorted_array2_quick)

print()

# print("Array 3 na bubble sort:", sorted_array3_bubble)
# print("Array 3 na selection sort:", sorted_array3_selection)
# print("Array 3 na insertion sort:", sorted_array3_insertion)
# print("Array 3 na merge sort:", sorted_array3_merge)
# print("Array 3 na quick sort:", sorted_array3_quick)

```
