```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

mall_customers_df = pd.read_csv("../../sqllite/mall_customers.csv")
X = mall_customers_df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)

# Get the cluster centers
cluster_centers = kmeans.cluster_centers_

# Get the labels for each data point
labels = kmeans.labels_

# Calculate the inertia (intra-cluster distance)
inertia = kmeans.inertia_

# Plot the clusters
plt.figure(figsize=(8, 6))

# Plot data points
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)

# Plot cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=200, label='Cluster Centers')

# Set labels and title
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('KMeans Clustering of Mall Customers Dataset')

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Print inertia (intra-cluster distance)
print("Inertia (Intra-cluster distance):", inertia)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

mall_customers_df = pd.read_csv("../../sqllite/mall_customers.csv")

X = mall_customers_df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)

# Get the cluster centers
cluster_centers = kmeans.cluster_centers_

# Predict cluster labels for new data points
new_data = [[60, 50], [80, 20], [40, 80]]
new_data_scaled = scaler.transform(new_data)
predicted_labels = kmeans.predict(new_data_scaled)
print("Predicted labels for new data points:", predicted_labels)

plt.figure(figsize=(8, 6))

# data points
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', s=50, alpha=0.5)

# cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=200, label='Cluster Centers')

# new data points
plt.scatter(new_data_scaled[:, 0], new_data_scaled[:, 1], c=predicted_labels, cmap='viridis', marker='o', s=100, edgecolors='black', label='New Data')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('KMeans Clustering of Mall Customers Dataset')

# plt.legend()

plt.grid(True)
plt.show()
```
