```python
import pandas as pd
from sklearn.cluster import KMeans
import math
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import StandardScaler
warnings.simplefilter('ignore')
import sqlite3
```

<h2>Data inlezen</h2>

```python
conn = sqlite3.connect('..\\..\\sqllite\\go_sales.sqlite')

sales_branch = pd.read_sql_query("SELECT * FROM sales_branch", conn)
country = pd.read_sql_query("SELECT * FROM country", conn)
```

<h2>Merge</h2>

```python
merged_sales = pd.merge(sales_branch, country, on='COUNTRY_CODE', how='inner')
merged_sales
```

<h2>Clean up</h2>

```python
merged_sales.drop(columns=['SALES_BRANCH_CODE', 'TRIAL888', 'COUNTRY_CODE', 'LANGUAGE', 'CURRENCY_NAME', 'TRIAL879'], inplace=True)

merged_sales
```

<h2>Dummy-fy</h2>

```python
merged_sales_dummies = pd.get_dummies(merged_sales)

merged_sales_dummies
```
